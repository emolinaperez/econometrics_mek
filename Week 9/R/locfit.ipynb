# Local Regression, Likelihood and Density Estimation

```r
install.packages('locfit')
library(locfit)
```

## Load Data (e.g., ethanol)

```r
data(ethanol)
head(ethanol)
```

## Visualize data

```r
plot(ethanol$E, ethanol$NOx, main = "Dataset ethanol")
```

## Fit Models with Varying Bandwidths

```r
# Bandwidth 0.1
fit1 <- locfit(NOx ~ E, data=ethanol, alpha=0.1)

# Bandwidth 0.5
fit2 <- locfit(NOx ~ E, data=ethanol, alpha=0.5)

# Bandwidth 1
fit3 <- locfit(NOx ~ E, data=ethanol, alpha=1)
```

## Visualize Results

```r
plot(ethanol$E, ethanol$NOx, main = "Dataset ethanol")
lines(fit1, col="red")
lines(fit2, col="blue")
lines(fit3, col="green")

legend("topright", legend=c("BW=0.1", "BW=0.5", "BW=1"),
       col=c("red", "blue", "green"), lty=1)
```

## Evaluate Fit with Cross-Validation

```r
# Compare Generalized Cross-Validation (GCV) scores
gcv(fit1)
gcv(fit2)
gcv(fit3)
```
